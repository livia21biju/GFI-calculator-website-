```python
import os
import pandas as pd

# Define directory path
directory_path = r"C:/Users/livia.biju/Desktop/code/Automate report"

# Define keywords for categorization
keywords = ["cii", "AE Boiler"]

# List all files in the directory
files = os.listdir(directory_path)

# Filter files that contain any of the keywords
filtered_files = [f for f in files if any(keyword.lower() in f.lower() for keyword in keywords)]

# Read files and assign DataFrame names dynamically
for file in filtered_files:
    file_path = os.path.join(directory_path, file)

    #Identify the keyword in the filename
    df_name = next((kw.lower().replace(" ", "_") for kw in keywords if kw.lower() in file.lower()), None)

    if df_name:
        #Read data
        if file.endswith(".csv"):
            df = pd.read_csv(file_path)
        elif file.endswith((".xlsx", ".xlsm")): # Modern Excel formats
            df = pd.read_excel(file_path, engine='openpyxl')
        elif file.endswith(".xls"): # Older Excel format
            df = pd.read_excel(file_path, engine='xlrd')
        else:
            print(f"Skipping unsupported file format: {file}")
            continue # Skip to the next file

        #Assign DataFrame to a variable dynamically
        globals()[df_name] = df
        print(f"Extracted data from: {file}  Stored as '{df_name}'")
```

```python
# Import matplotlib
import matplotlib.pyplot as plt
import os # Import os module for path manipulation

def plot_donut_chart(average_vpr, filename="vpr_donut_chart.png"):
    """
    Generates a donut chart representing the Vessel Performance Rating (VPR)
    and saves it as a PNG image.

    Args:
        average_vpr (float): The average VPR value (expected to be between 0 and 100).
        filename (str, optional): The desired path and name for the output PNG file.
                                   Defaults to "vpr_donut_chart.png".
    """
    # --- Input Validation (Optional but Recommended) ---
    if not 0 <= average_vpr <= 100:
        print(f"Warning: average_vpr ({average_vpr}) is outside the expected 0-100 range.")
        # Clamp the value to ensure correct portion calculation
        average_vpr = max(0, min(average_vpr, 100))

    # --- Data Preparation ---
    # Define the portion filled (based on the average value) and the remaining part
    filled_portion = average_vpr / 100.0 # Ensure float division
    remaining_portion = 1.0 - filled_portion

    # Handle edge case where filled_portion is 0 or 1 to avoid tiny slivers if possible
    if filled_portion == 0:
        chart_data = [1.0] # Only remaining portion
        colors = ['#D3D3D3'] # Use a single color for remaining (e.g., light gray)
    elif remaining_portion == 0:
         chart_data = [1.0] # Only filled portion
         colors = ['#002060'] # Use a single color for filled
    else:
        chart_data = [filled_portion, remaining_portion]
        colors = ['#002060', '#FFFFFF'] # Dark Blue for filled, Light Gray for remaining

    # --- Chart Creation ---
    # Create the figure and axes objects
    fig, ax = plt.subplots(figsize=(6, 6)) # Use fig, ax for better control

    # Create the pie chart (wedges)
    wedges, _ = ax.pie(chart_data, colors=colors, startangle=90,
                       wedgeprops={'width': 0.35, 'edgecolor': 'white'}) # Slightly wider width

    # Add the average value text in the center
    ax.text(0, 0, f"{average_vpr:.1f}%", ha='center', va='center', fontsize=20, fontweight='bold', color='#002060') # Match color, add %

    # Ensure the chart is circular
    ax.set_aspect('equal') # Use ax.set_aspect

    # Add a title to the axes
    #ax.set_title("Vessel Performance Rating", fontsize=14, pad=20) # Add padding

    # --- Saving the Plot ---
    # Use fig.savefig() for more control
    # Recommended parameters:
    #   dpi: Dots Per Inch (resolution) - 300 is good for general use
    #   bbox_inches='tight': Reduces extra whitespace and prevents labels/title from being cut off
    #   transparent=True: Optional, for a transparent background
    try:
        fig.savefig(filename, dpi=300, bbox_inches='tight')
        print(f"Chart saved successfully as '{os.path.abspath(filename)}'") # Show absolute path
    except Exception as e:
        print(f"Error saving chart to '{filename}': {e}")

    # --- Displaying the Plot (Optional) ---
    # plt.show() # Uncomment this line if you want to display the plot interactively as well

    # --- Clean up ---
    plt.close(fig) # Close the figure to free up memory, especially if calling the function in a loop
# Call the function to plot and save
plot_donut_chart(average_vpr, filename="VPR.png")

# Example with a different filename
# plot_donut_chart(95.2, filename="high_vpr_rating.png")

# Example with value 0
# plot_donut_chart(0.0, filename="zero_vpr_rating.png")

# Example with value 100
# plot_donut_chart(100.0, filename="perfect_vpr_rating.png")
```

```python
import pandas as pd

def create_vpr_table(num_files, vpr_datasets):
    """
    Create a pandas DataFrame dynamically based on num_files.
    
    :param num_files: Number of available quarter files (1 to 4).
    :param vpr_datasets: Dictionary containing datasets {1: vpr1, 2: vpr2, ..., 4: vpr4}.
    :return: pandas DataFrame or 0 if num_files is 0.
    """
    if num_files == 0:
        return 0  # No files exist

    # Determine the reference dataset for vessel names
    vessel_data = vpr_datasets.get(num_files)  # Uses vpr4 if num_files=4, vpr3 if num_files=3, etc.
    if vessel_data is None:
        return 0  # No valid dataset found

    # Initialize table with Vessel Names from the selected dataset
    df = pd.DataFrame()
    df["Vessel Name"] = vessel_data["Vessel"]  # Take vessel names from the highest available dataset

    # Add VPR data for each available quarter
    for q in range(1, num_files + 1):
        vpr_df = vpr_datasets.get(q)  # Get vpr1, vpr2, etc.
        if vpr_df is not None:
            df = df.merge(
                vpr_df[["Vessel", "Vessel Performance Rating"]],
                left_on="Vessel Name",
                right_on="Vessel",
                how="left"
            ).drop(columns=["Vessel"])
            df.rename(columns={"Vessel Performance Rating": f"Q{q} VPR(%)"}, inplace=True)

    return df
# Generate the table
vpr_table = create_vpr_table(num_files, vpr_datasets)

# Print the table
print(vpr_table)
```

```python
import pandas as pd

def highlight_vpr(val):
    """Apply background color based on VPR value"""
    # Check if the value is numeric (int or float)
    # and not NaN, as NaN would cause an error in comparisons
    if isinstance(val, (int, float)) and not pd.isna(val):
        if val >= 70:
            return 'background-color: lightgreen; color: black;'
        elif 40 <= val < 70:
            return 'background-color: orange; color: black;'
        else: # val < 40
            return 'background-color: red; color: white;'
    # Return empty string for non-numeric values or NaN to apply no specific style
    return ''

def style_vpr_table(df):
    """Apply styling to the VPR table with grid lines"""
    # Ensure the DataFrame passed is not empty and has columns
    if df.empty or len(df.columns) == 0:
        print("DataFrame is empty or has no columns. Returning original DataFrame.")
        return df

    # Identify numeric columns for formatting and styling
    # Exclude the first column if it's meant to be categorical (like 'Category')
    numeric_cols_to_style = df.columns[1:] if len(df.columns) > 1 else []

    # Start styling
    styled_df = df.style

    # Apply cell-wise styling using .map() for numeric columns
    # Check if numeric_cols_to_style is not empty before applying map
    if numeric_cols_to_style.size > 0 : #.size checks if the Index is empty
        styled_df = styled_df.map(highlight_vpr, subset=numeric_cols_to_style)
        # Format numeric columns (only if they exist)
        styled_df = styled_df.format("{:.1f}", subset=numeric_cols_to_style, na_rep="-") # na_rep handles NaN display

    # Style the first column (e.g., 'Category')
    if len(df.columns) > 0:
        styled_df = styled_df.set_properties(
            **{'font-weight': 'bold', 'color': 'darkblue'},
            subset=pd.IndexSlice[:, df.columns[0]]
        )

    # Style the other (numeric) columns' text color (only if they exist)
    if numeric_cols_to_style.size > 0:
        styled_df = styled_df.set_properties(
            **{'color': 'darkblue'},
            subset=pd.IndexSlice[:, numeric_cols_to_style]
        )

    # Hide the index
    styled_df = styled_df.hide(axis='index')

    # Apply table-wide styles for headers and cells (grid lines)
    styled_df = styled_df.set_table_styles([
        {'selector': 'th', 'props': [
            ('border', '1px solid black'),
            ('background-color', 'darkblue'),
            ('color', 'white'),
            ('font-weight', 'bold'),
            ('text-align', 'center') # Center align headers
        ]},
        {'selector': 'td', 'props': [
            ('border', '1px solid black'),
            ('text-align', 'center') # Center align data cells
        ]}
    ])
    return styled_df

# Apply styling to your existing vpr_table
styled_table = style_vpr_table(vpr_table.copy()) # Use .copy() to avoid modifying the original DataFrame
# To display in a Jupyter Notebook or similar environment:
styled_table

# If you want to save it to HTML:
# styled_table.to_html('styled_vpr_table.html', escape=False)
# print("Styled VPR table saved to HTML.")

# For script execution, you might want to see the HTML or just confirm it runs
if hasattr(styled_table, 'to_html'): # Check if styling was applied
    print("Styling applied. To view, display 'styled_table' in a Jupyter environment or save to HTML.")
    # Example of saving to HTML
    try:
        html_output = styled_table.to_html(escape=False)
        with open('styled_vpr_table.html', 'w') as f:
            f.write(html_output)
        print("Styled VPR table saved to styled_vpr_table.html")
    except Exception as e:
        print(f"Error saving to HTML: {e}")
else:
    print("Styling could not be applied (e.g., DataFrame was empty).")
    
```

```python
import pandas as pd
from datetime import datetime

def style_table(table, report_type):
    """Apply styling to the table, including conditional formatting for savings."""

    # Function to style the Savings column
    def color_savings(val):
        """Applies background and text color based on the numeric value of savings."""
        try:
            # Attempt to convert the value to a float for comparison
            numeric_val = float(val)
            if numeric_val > 0:
                return 'background-color: #90EE90; color: black'  # Light green for positive
            elif numeric_val < 0:
                return 'background-color: #FF6347; color: white'  # Tomato red for negative
            else: # numeric_val == 0
                return 'background-color: #FFFFFF; color: black'  # White for zero
        except (ValueError, TypeError):
            # Handles cases like 'No Valid Logs', None, or other non-numeric strings
            return 'background-color: #FFFFFF; color: black'  # White for non-numeric

    # Determine the Savings column name based on report_type
    current_year = datetime.now().strftime('%y')  # e.g., '25' for 2025
    savings_col = None
    if report_type == 'Q2':
        savings_col = f'Savings Q1-Q2/{current_year}'
    elif report_type == 'Q3':
        savings_col = f'Savings Q2-Q3/{current_year}'
    elif report_type == 'Q4':
        savings_col = f'Savings Q3-Q4/{current_year}'
    # Add more conditions if other report types exist or provide a default

    # Start styling
    styled_table = table.style

    # Apply general properties to all cells
    styled_table = styled_table.set_properties(**{
        'background-color': '#F5F5F5',  # Light gray background for cells
        'color': 'black',                # Black text for cells
        'border-color': '#D3D3D3',       # Light gray borders
        'border-style': 'solid',
        'border-width': '1px',
        'text-align': 'center',
        'font-size': '12px'
    })

    # Apply styles to table headers
    styled_table = styled_table.set_table_styles([
        {'selector': 'th',
         'props': [('background-color', '#1E90FF'),  # Dark blue header background
                   ('color', 'white'),               # White header text
                   ('font-weight', 'bold'),          # Bold headers
                   ('text-align', 'center'),
                   ('border-color', '#D3D3D3'),
                   ('border-style', 'solid'),
                   ('border-width', '1px'),
                   ('font-size', '14px')]}
    ])

    # Conditionally apply styling to the savings column if it exists
    if savings_col and savings_col in table.columns:
        # Use .map() for element-wise styling on the specific savings column
        styled_table = styled_table.map(color_savings, subset=[savings_col])
    else:
        print(f"Warning: Savings column '{savings_col}' not found in table. Skipping savings-specific styling.")

    # Format all numeric columns to 2 decimal places.
    # Styler.format() can take a dictionary to format specific columns,
    # or a general formatter for all.
    # It's often better to identify numeric columns if mixing types,
    # but for general precision on numbers, this is okay.
    # Using na_rep to represent NaN values cleanly.
    styled_table = styled_table.format(precision=2, na_rep='-')

    # Hide the index for display
    styled_table = styled_table.hide(axis='index')

    return styled_table
# Replace NaN values with "No Valid Logs" in all columns except 'Name' and 'IMO'
for col in table_data1.columns:
    if col not in ['Name', 'IMO']:  # Skip 'Name' and 'IMO' columns
        table_data1[col] = table_data1[col].fillna('No Valid Logs')

# Assuming tableM and Report_type are defined (sample definitions above)
if 'table_data1' in locals() and 'Report_type' in locals():
    styled_table_data1 = style_table(table_data1.copy(), Report_type) # Use .copy()

    # To display in a Jupyter Notebook or similar environment:
    # styled_tableM
    print("Styled table 'styled_table_data1' created. Display in Jupyter or save to Excel.")

    # Save to Excel without index
    try:
        # Note: Styler.to_excel() handles the styling export to Excel.
        styled_table_data1.to_excel('styled_table_data1.xlsx', engine='openpyxl', index=False)
        print("Styled table saved to 'styled_table_data1.xlsx'")
    except Exception as e:
        print(f"Error saving to Excel: {e}")
else:
    print("Please define 'table_data1' and 'Report_type' to run the styling example.")
                 
```

```python
import pandas as pd
from datetime import datetime

def style_table(table, report_type):
    """Apply styling to the table, including conditional formatting for savings."""

    # Function to style the Savings column
    def color_savings(val):
        """Applies background and text color based on the numeric value of savings."""
        try:
            # Attempt to convert the value to a float for comparison
            numeric_val = float(val)
            if numeric_val > 0:
                return 'background-color: #90EE90; color: black'  # Light green for positive
            elif numeric_val < 0:
                return 'background-color: #FF6347; color: white'  # Tomato red for negative
            else: # numeric_val == 0
                return 'background-color: #FFFFFF; color: black'  # White for zero
        except (ValueError, TypeError):
            # Handles cases like 'No Valid Logs', None, or other non-numeric strings
            return 'background-color: #FFFFFF; color: black'  # White for non-numeric

    # Determine the Savings column name based on report_type
    current_year = datetime.now().strftime('%y')  # e.g., '25' for 2025
    savings_col = None
    if report_type == 'Q2':
        savings_col = f'Savings Q1-Q2/{current_year}'
    elif report_type == 'Q3':
        savings_col = f'Savings Q2-Q3/{current_year}'
    elif report_type == 'Q4':
        savings_col = f'Savings Q3-Q4/{current_year}'
    # Add more conditions if other report types exist or provide a default

    # Start styling
    styled_table = table.style

    # Apply general properties to all cells
    styled_table = styled_table.set_properties(**{
        'background-color': '#F5F5F5',  # Light gray background for cells
        'color': 'black',                # Black text for cells
        'border-color': '#D3D3D3',       # Light gray borders
        'border-style': 'solid',
        'border-width': '1px',
        'text-align': 'center',
        'font-size': '12px'
    })

    # Apply styles to table headers
    styled_table = styled_table.set_table_styles([
        {'selector': 'th',
         'props': [('background-color', '#1E90FF'),  # Dark blue header background
                   ('color', 'white'),               # White header text
                   ('font-weight', 'bold'),          # Bold headers
                   ('text-align', 'center'),
                   ('border-color', '#D3D3D3'),
                   ('border-style', 'solid'),
                   ('border-width', '1px'),
                   ('font-size', '14px')]}
    ])

    # Conditionally apply styling to the savings column if it exists
    if savings_col and savings_col in table.columns:
        # Use .map() for element-wise styling on the specific savings column
        styled_table = styled_table.map(color_savings, subset=[savings_col])
    else:
        print(f"Warning: Savings column '{savings_col}' not found in table. Skipping savings-specific styling.")

    # Format all numeric columns to 2 decimal places.
    # Styler.format() can take a dictionary to format specific columns,
    # or a general formatter for all.
    # It's often better to identify numeric columns if mixing types,
    # but for general precision on numbers, this is okay.
    # Using na_rep to represent NaN values cleanly.
    styled_table = styled_table.format(precision=2, na_rep='-')

    # Hide the index for display
    styled_table = styled_table.hide(axis='index')

    return styled_table
# Replace NaN values with "No Valid Logs" in all columns except 'Name' and 'IMO'
for col in table_data111.columns:
    if col not in ['Name', 'IMO']:  # Skip 'Name' and 'IMO' columns
        table_data111[col] = table_data111[col].fillna('No Valid Logs')

# Assuming tableM and Report_type are defined (sample definitions above)
if 'table_data11' in locals() and 'Report_type' in locals():
    styled_table_data111 = style_table(table_data111.copy(), Report_type) # Use .copy()

    # To display in a Jupyter Notebook or similar environment:
    # styled_tableM
    print("Styled table 'styled_table_data1' created. Display in Jupyter or save to Excel.")

    # Save to Excel without index
    try:
        # Note: Styler.to_excel() handles the styling export to Excel.
        styled_table_data111.to_excel('styled_table_data11.xlsx', engine='openpyxl', index=False)
        print("Styled table saved to 'styled_table_data11.xlsx'")
    except Exception as e:
        print(f"Error saving to Excel: {e}")
else:
    print("Please define 'table_data111' and 'Report_type' to run the styling example.")
        
```
